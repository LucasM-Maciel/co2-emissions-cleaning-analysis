```python
df_columns = df_no_nan.columns
emission_value_columns = df_columns[4:36]
filter_1 = df_no_nan[emission_value_columns].all(axis=1) ==0
df_first_cleaning = df_no_nan[~filter_1]
df_first_cleaning = df_first_cleaning.reset_index(drop=True)
df_first_cleaning
```

```python
df_first_cleaning.duplicated().sum()
```

```python
df_sec_cleaning = df_first_cleaning.drop(columns=["country_code","Indicator Name"])
df_sec_cleaning
```

```python
df_sec_cleaning["Total Emission Value"] = df_sec_cleaning[emission_value_columns].sum(axis=1)
df_sorted = df_sec_cleaning.sort_values("Total Emission Value", ascending = False)
df_organized = df_sorted.reset_index(drop=True)
df_organized
```

```python
df_percent = df_organized[emission_value_columns].pct_change(axis=1)
YoY_spike = (df_percent > 111) | (df_percent < -0.991)
countries_with_spikes = df_organized.loc[YoY_spike.any(axis=1),"Country"]
countries_with_spikes
```

```python
rows, columns = np.where(YoY_spike)

year_columns_array = np.array(emission_value_columns)
spike_year = year_columns_array[columns]
previous_year = np.where(columns > 0, year_columns_array[columns - 1], None)

emission_values = df_organized[emission_value_columns].values
previous_value = np.where(columns > 0, emission_values[rows, columns - 1], np.nan)
spike_value = emission_values[rows, columns]

df_spikes = pd.DataFrame({
    "Country": df_organized["Country"].iloc[rows].values,
    "Previous_Year": previous_year,
    "Year": spike_year,
    "Previous_value": previous_value,
    "Spike_Value": spike_value,
})
df_spikes["YoY_pct_change_%"] = (
    ((df_spikes["Spike_Value"] - df_spikes["Previous_value"]) / df_spikes["Previous_value"]) * 100
).map(lambda x: f"{x:,.1f}%" if np.isfinite(x) else None)

df_spikes
```

## Context hypothesis (1991–1993)

Mauritania shows an abrupt near-total drop in emissions followed by an extreme rebound between 1991 and 1993.

A plausible interpretation is that this pattern reflects a combination of external shocks and structural changes during the early 1990s. Reports from that period mention severe droughts affecting agriculture and the economy, as well as a major political transition around 1991 that may have influenced economic activity and reporting practices.

Because this analysis does not validate the figures against official emissions inventories, I treat the 1991–1993 changes as **potential anomalies** and keep them in the dataset while flagging them for caution in downstream interpretations.
